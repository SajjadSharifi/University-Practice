```python
#بسم الله الرحمن الرحیم
#محمد سجاد شریفی پناه
import jupyturtle
import math
```

```python
def rectangle(width: int, length: int) -> None:
    try:
        jupyturtle.make_turtle()
        for i in range(2):
            jupyturtle.left(90)
            jupyturtle.forward(width)
            jupyturtle.left(90)
            jupyturtle.forward(length)
    except Exception as exception:
        raise exception(" something went wrong!! ")
    

def rhombus(length: int, angle: int) -> None:
    try:
        jupyturtle.make_turtle()
        for i in range(2):
            jupyturtle.forward(length)
            jupyturtle.left(angle)
            jupyturtle.forward(length)
            jupyturtle.left(180-angle)
    except Exception as exception:
        raise exception(" something went wrong!! ")

def parallelogram(length1: int, length2: int, angle: int) -> None:
    try:
        jupyturtle.make_turtle()
        for i in range(2):
           jupyturtle.forward(length1)
           jupyturtle.left(angle)
           jupyturtle.forward(length2)
           jupyturtle.left(180-angle)
    except Exception as exception:
        raise exception(" something went wrong!! ")

def petal(radius: int,  angle: int):
    try:
        for i in range(2):
            jupyturtle.arc(radius,  angle)
            jupyturtle.left(180-angle)
    except Exception as exception:
        raise exception(" something went wrong!! ")

def flower(n: int, radius: int, angle: int):
    try:
        for i in range(n):
            petal(radius,  angle)
            jupyturtle.left(360.0/n)
    except Exception as exception:
        raise exception(" something went wrong!! ")

def spiral(length, angle):
    try:
        turtle = jupyturtle.make_turtle(delay=0)
        for _ in range(100):
            jupyturtle.forward(length)
            jupyturtle.right(angle)
            length += 1  # Increase the length for each segment
    except Exception as exception:
        raise exception(" something went wrong!! ")
 
        
def circular_spiral(radius, angle):
    try:
        turtle = jupyturtle.make_turtle(delay=0, width=500, height=500)
        rotations = 5
        distance = 2 * math.pi * radius / 360  # Calculate the distance for each degree of rotation
        for _ in range(rotations * 360):
            jupyturtle.forward(distance)
            jupyturtle.left(angle)
            radius += 0.1  # Increase the radius for each segment
            distance = 2 * math.pi * radius / 360  # Recalculate the distance for the updated radius
    except Exception as exception:
        raise exception(" something went wrong!! ")
   
   
```
